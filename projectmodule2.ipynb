Untuk mengatasi missing value yang terdapat pada kolom diatas, beberapa metode dapat dilakukan. Yang pertama adalah menghapus data(MAR/Missing at Random) yang kedua adalah mengimputasi data kosong dengan nilai mean atau median bergantung dari distribusi data yang terbentuk(MAR/Missing at Random) dan yang ketiga adalah mengimputasi nilai dengan data yang berhubungan dengan kolom tersebut (missing not at random).

```python
df.head()
df_num=['Age','Flight Distance','Departure Delay in Minutes','Arrival Delay in Minutes']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5,5))
for i in df_num:
    fig,ax=plt.subplots(1,2)
    sns.boxplot(df[i],ax=ax[0]).set(title=i)  
    ax[1]=plt.hist(df[i],bins=100,edgecolor='black',linewidth=0.5)
    plt.title(i)
    plt.tight_layout()
```

Pada gambar diatas beberapa kolom dengan tipe data numerikal memiliki keberadaan outliers yang cukup signifikan. Beberapa diantaranya adalah kolom flight distance,departure delay in minutes, dan arrival delay in minutes. 
Untuk mengatasi outliers tersebut terdapat beberapa metode yang dapat dilakukan seperti :
1. Trimming data outlier
2. Membiarkan data apabila data outliers masih dikarenakan data masih memungkinkan terjadi pada dunia nyata.

```python
# Mencari ratio data outliers dengan keseluruhan.
for i in df_num:
    IQR=np.percentile(df[i],75)-np.percentile(df[i],25)
    ba=1.5*IQR+np.percentile(df[i],75)
    bb=np.percentile(df[i],25)-1.5*IQR
    proporsi=len([i for i in df[i] if i>ba or i<bb])/df['Arrival Delay in Minutes'].count()*100
    print(f' Jumlah Proporsi Data {i} outliers dengan Kesulurahan data :{proporsi} %')
```

Pada hasil proporsi dapat terlihat hasil data yang masih cukup baik untuk dihapus adalah `Flight Distance` namun kita perlu menelaah terlebih dahulu apakah data tersebut berada pada nilai yang tidak semestinya atau masih dalam batas wajar.

## Inflight Wifi Service Profil pada Flight Distance

```python
sns.boxplot(x=df['Inflight wifi service'],y=df['Flight Distance'])
```

Pada gambar diatas rata rata pada jarak flight distance yang didasari oleh penilaian pelayanan wifi 1 dan 5 masih cukup sulit dibedakan oleh boxplot.Dikarenakan hal itu diperlukan hipotesis testing. Hal ini dilakukan untuk mengetahui rata rata flight distance pada setiap rating pelayanan wifi. Penilaian 1 dan 5 dipilih menjadi sampel dikarenakan memiliki perbedaan kualitas yang cukup jauh. Untuk penilaian 0 tidak digunakan dikarenakan kemungkinan besar beberapa penumpang memilih untuk tidak mengisi atau "abstain" sehingga tidak merepresentasikan penilaian yang buruk. Uji Hipotesis:
1. Apakah rata rata flight distance pada penilaian 1 memiliki rata rata jarak penerbangan yang lebih jauh atau tidak dibandingkan dengan penilaian 5 pada pelayanan wifi?

```python
import statsmodels.api as sm
import pylab as py
sm.qqplot(df['Flight Distance'][df['Inflight wifi service']==1],line='45')
sm.qqplot(df['Flight Distance'][df['Inflight wifi service']==5],line='45')
py.show()
```

Melalui dua grafik diatas dapat kita lihat bahwa kedua sampel yang akan kita uji tidak terdistribusi secara normal, sehingga kita memerlukan pengujian non parametrik untuk kedepannya. Pengujian yang dipilih ada mannwhitney untuk mengukur perbedaan dua rata rata dari dua sampel yang independen satu sama lain. Output yang dihasilkan adalah besaran p-value yang menjadi acuan untuk menilai rata rata jarak tempuh pesawat yang lebih besar pada penilaian yang buruk dan paling baik.

```python
flight_1=df['Flight Distance'][df['Inflight wifi service']==1]
flight_5=df['Flight Distance'][df['Inflight wifi service']==5]
```

```python
from scipy.stats import mannwhitneyu
hypo_flight=mannwhitneyu(flight_1,flight_5,alternative='less')
```

```python
hypo_flight
```

Melalui testing diatas, nilai p-value lebih kecil dibandingkan alpha (0.05) sehingga kita perlu menolak h0 dan menerima hipotesis alternatif yakni rata rata jarak penerbangan pada penilaian pelayanan wifi rating 1 lebih rendah dibandingkan pada penilaian pelayanan wifi dengan rating 5.Dengan kata lain rata rata penerbangan jarak pendek memiliki kualitas wifi yang buruk dalam pesawat, hal ini perlu diperhatikan dan tidak boleh luput dikarenakan apabila kualitas wifi yang baik akan meningkatkan keamanan pesawat juga dikarenakan mencegah adanya penggunaan jaringan seluler di dalam kabin yang dapat merusak radar pesawat.
